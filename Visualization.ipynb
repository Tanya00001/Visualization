# VISUALIZATION

## In visualization there are two things:

## 1. Univariate - looking at a single variable and visualizing the data
###                 eg:  *for int or float type data which is univariate, we uses - HISTOGRAM(there                                  is sort of bin or range). In this bars are together.*
###                      *for string type data which is univariate, we uses - BAR CHART. Here bars                                      are separated.*

## 2. Bivariate - looking at two d/f variable and visualizing the data
###                 eg: *for int or float type data which is bivariate, we uses - SCATTER PLOT. Here                                 we fits a line to see the trend.*
###                        *for string with int/float, we uses - BOX PLOT(in median fashion or can                                        say in percentile)*

#### Here visualization makes sense. if we need to visualize with more variables or parameters means more than that of bivariate, there ML models will work in a better efficient way because even if we will plot that, still it would be difficult to understand what is actually going on.

## *Bi-variate Analysis*

### Que: Does the price of the house depend on the LotArea?

```python
sns.regplot(x = "LotArea", y = "SalePrice", data = housing)

# regplot is for scatter plot, it will fit a regression line along with this.
```

### Removing outliers

```python
# firstly we will check,
# are outliers skewing the relationship?
# if yes, redraw relationship after removing very large values

# checking..
housing["LotArea"].quantile([0.5,0.95,0.99])

# we can check percentiles values to understand from which percentile outliers are starting
```

```python
# here we can see the values on 0.5(median), 0.95, and 0.99 percentile. As on the 0.99 the value is quite larger than at 0.5, means there are some outliers.
# so for that we are removing every value which is above 0.95 percentile..

housing_sub = housing.loc[housing["LotArea"] < housing["LotArea"].quantile(0.95)]
```

```python
# now, if we will plot this..

sns.regplot( x = "LotArea", y = "SalePrice", data = housing_sub)
```

#### so now it is much more without outliers, considering the data less than 0.95 percentile
#### now we can answer by looking at this..

### Ans: yes, as the LotArea increasing housing price is also increasing.

# Plotting multiple graphs

### Que: Analyse the relationship b/w SalesPrice and all "Square Feet(SF)" related columns

```python
# here we are selecting columns whose names are having this SF and storing them in sf_cols.

sf_cols = [col_name for col_name in housing.columns if "SF" in col_name]   # list comprehension
```

```python
# No. of columns in which SF(Square Feet) is there..

len(sf_cols)
sf_cols
```

#### Now, if we want to plot multiple graphs but in a single canvas.

```python
# here if we want to plot 9 graphs in a single graph, 
# here is 9 SF data and for each SF data we want to plot the SalePrice, because probably we want to check every individual SF area impact on SalePrice.


# firstly, we will look at the structure of multiple plot within a single graph.
fig, axs = plt.subplots(nrows = 3, ncols = 3, figsize = (10,10))
```

```python
#### a 3*3 matrix like structure
#### if we want to see a particular subgraph then we have to say that first one is at (0,0), second at (0,1) and so on..

fig, axs = plt.subplots( nrows = 3, ncols = 3, figsize = (10,10))
for i in range(0, len(sf_cols)):
    rows = i // 3
    cols = i % 3
    print(rows,cols)
```

```python
# Now, we will plot all SF columns and SalePrice on this graph..


fig, axs = plt.subplots(nrows = 3, ncols = 3, figsize = (10,10))
for i in range(0, len(sf_cols)):
    rows = i // 3
    cols = i % 3
    ax = axs[rows,cols]
    plot = sns.regplot(x = sf_cols[i], y = "SalePrice", data = housing, ax = ax)
    sns.set_theme(color_codes = True , style = "darkgrid", palette = "rainbow")
```
